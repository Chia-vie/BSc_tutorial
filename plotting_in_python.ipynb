```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
myarr = np.array([1,2,3,4,5])
otherarr = np.array([2.5, 3.3, 1.5, 3.1, 4.2])
```

```python
plt.plot(myarr, otherarr)
```

```python
plt.scatter(myarr, otherarr)
```

```python
plt.plot(myarr, otherarr)
plt.xlabel('x')
plt.ylabel('y')
```

### Style

```python
plt.plot(myarr, otherarr, c='gold')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
# larger tick labels
```

### Saving a figure

```python
plt.plot(myarr, otherarr, c='gold')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
plt.savefig('myplot.png')
```

### Fill between

```python
plt.plot(myarr, otherarr, c='gold')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
plt.fill_between(1.5, otherarr)

# anschauen!!
```

```python
errors = np.array([0.1, 0.2, 0.5, 0.3, 0.9])
plt.scatter(myarr, otherarr, c='gold')
plt.errorbar(myarr, otherarr, errors, fmt='o')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
# anschauen!!
```
