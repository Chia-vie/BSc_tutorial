### Creating a virtual evironment
Virtual environments are very useful when it comes to managing Python packages. 
Whenever you start a new project (such as your thesis project) you want to create a project environment.
This way you can ensure that your project will still run without complications in the future and you can avoid module version conflicts.

in terminal or command prompt type:
```
python -m venv myvenv
```

here myvenv is the name of the new virtual environment and can be changed to anything you like.

If you receive an error type:

```
pip install venv
```
or 

```
pip3 install venv
```

and then repeat the previous step.

To activate your new environment in Linux/Mac terminal type:

```
source myvenv/bin/activate
```

in Windows cmd type:

```
myvenv\Scripts\activate.bat
```
(note the backslashes above)

to exit the environment type:

```
deactivate
```

### Simple plots:

````python
import matplotlib.pyplot as plt
import numpy as np
````

Create some data

````python
myarr = np.array([1,2,3,4,5])
otherarr = np.array([2.5, 3.3, 1.5, 3.1, 4.2])
````

Plot data

````python
plt.plot(myarr, otherarr)
plt.show() # not needed in jupyter notebook when only code for one plot is in one cell 
````

````python
plt.scatter(myarr, otherarr)
````

### Add axis labels + title

````python
plt.plot(myarr, otherarr)
plt.xlabel('x')
plt.ylabel('y')
plt.title('This is my cool plot')
````

### Change fontsizes + tick labels + colour

````python
plt.plot(myarr, otherarr, c='gold')
plt.xlabel('huge axis', fontsize=20)
plt.ylabel('tiny axis', fontsize=10)
plt.title('This is my cool plot', fontsize=30)
xticks = plt.xticks(fontsize=15)
yticks = plt.yticks(fontsize=5)
# larger tick labels
````

### Remember: Always make your plots readable!

### Plotting and labeling multiple datasets

````python
othervals = np.array([2.1, 3.2, 1.7, 2.8, 3.9])
plt.plot(myarr, otherarr, c='black', label='my values')
plt.plot(myarr, othervals, c='gold', label="someone else's values")
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
xticks = plt.xticks(fontsize=15)
yticks = plt.yticks(fontsize=15)#
plt.legend()
````

### Saving a figure

````python
othervals = np.array([2.1, 3.2, 1.7, 2.8, 3.9])
plt.plot(myarr, otherarr, c='black', label='my values')
plt.plot(myarr, othervals, c='gold', label="someone else's values")
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
xticks = plt.xticks(fontsize=15)
yticks = plt.yticks(fontsize=15)#
plt.legend()
plt.savefig('myplot.png')
````

````python
errors = np.array([0.1, 0.2, 0.5, 0.3, 0.9])
plt.scatter(myarr, otherarr, c='gold')
plt.errorbar(myarr, otherarr, errors, fmt='o')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
````

### Fill between

````python
plt.plot(myarr, otherarr, c='gold')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool plot', fontsize=30)
plt.fill_between(myarr, otherarr, 1) # x values, y_max values, y_min values
````

### Histogram

````python
newarr = np.random.normal(0,0.25,1000) # random normal distribution around 1, 1000 values
plt.hist(newarr, bins=10) # number of bins = 10
plt.xlabel('x', fontsize=20)
plt.ylabel('count', fontsize=20)
plt.title('This is my cool histogram', fontsize=30)
xticks = plt.xticks(fontsize=15)
yticks = plt.yticks(fontsize=15)
plt.savefig('myhistogram.png')
````

````python
plt.hist(newarr, bins=5) # number of bins = 5
plt.hist(newarr, bins=10)
plt.hist(newarr, bins=20)
plt.xlabel('x', fontsize=20)
plt.ylabel('count', fontsize=20)
plt.title('This is my cool histogram', fontsize=30)
xticks = plt.xticks(fontsize=15)
yticks = plt.yticks(fontsize=15)
plt.savefig('myhistogram.png')
````

## Images

````python
values = np.random.rand(5, 5) # create a matrix with some random values
````

````python
print(values)
````

````python
plt.imshow(values, cmap='viridis') # number of bins = 5
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('This is my cool image', fontsize=30)
xticks = plt.xticks(fontsize=15)
yticks = plt.yticks(fontsize=15)
plt.colorbar(label='heat', orientation='vertical')
plt.savefig('myimage.png')
````

### Multiple plots in one

````python
fig, ax = plt.subplots(1, 3, figsize=(15, 5)) # number of rows, number of columns, figure size(width, height)
interpolation = ['nearest', 'bilinear', 'bicubic']
for i in range(3):
    im = ax[i].imshow(values, interpolation=interpolation[i])
    ax[i].set_title(interpolation[i]) # slightly different commands from before 
    ax[i].grid(True)
    ax[i].set_xlabel('x') 
    ax[i].set_ylabel('y')
fig.colorbar(im, ax=ax[:], orientation='vertical', label='heat', shrink=0.7)
# colorbar needs a mappable (im) and a subset of the ax object
# shrink colorbar it so it fits to the plot size
plt.show()
````
